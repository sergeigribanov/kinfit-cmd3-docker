```python
from ROOT import TFile, TCanvas, gDirectory, TLegend
```

```python
path_noisr = "kinfit_kskpi_run068705.root"
path_isr = "kinfit_kskpi_run068706.root"
path_isr_ph = "isr_run068706.root"
```

```python
fl_noisr = TFile.Open(path_noisr, 'read')
fl_isr = TFile.Open(path_isr, 'read')
fl_isr_ph = TFile.Open(path_isr_ph, 'read')
```

```python
cnv_noisr = TCanvas()
fl_noisr.kf_data.Draw("kf_mks>>kf_mks_noisr(512,0.4,0.6)", "kf_err==0")
cnv_noisr.Draw()
```

```python
cnv_isr = TCanvas()
fl_isr.kf_data.Draw("kf_mks>>kf_mks_isr(512,0.4,0.6)", "kf_err==0")
cnv_isr.Draw()
```

```python
cnv_isr_ph = TCanvas()
fl_isr_ph.kf_data.Draw("kf_mks>>kf_mks_isr(512,0.4,0.6)", "kf_err==0")
cnv_isr_ph.Draw()
```

```python
cnv = TCanvas()
fl_noisr.kf_data.Draw("kf_mks>>kf_mks_noisr(512,0.4,0.6)", "kf_err==0")
gDirectory.Get("kf_mks_noisr").SetStats(False)
gDirectory.Get("kf_mks_noisr").SetTitle("")
gDirectory.Get("kf_mks_noisr").GetXaxis().SetTitle("m_{#pi^{+}#pi^{#minus}} (GeV)")
gDirectory.Get("kf_mks_noisr").GetXaxis().SetTitleSize(0.06)
gDirectory.Get("kf_mks_noisr").GetXaxis().SetLabelSize(0.06)
gDirectory.Get("kf_mks_noisr").GetYaxis().SetLabelSize(0.06)
gDirectory.Get("kf_mks_noisr").GetYaxis().SetTitle("events")
gDirectory.Get("kf_mks_noisr").GetYaxis().SetTitleSize(0.06)
gDirectory.Get("kf_mks_noisr").SetLineColor(2)
gDirectory.Get("kf_mks_noisr").SetLineWidth(2)
gDirectory.Get("kf_mks_noisr").SetLineStyle(9)
fl_isr.kf_data.Draw("kf_mks>>kf_mks_isr(512,0.4,0.6)", "kf_err==0", "same")
gDirectory.Get("kf_mks_isr").SetLineColor(4)
gDirectory.Get("kf_mks_isr").SetLineWidth(2)
legend = TLegend(0.1,0.7,0.48,0.9)
legend.AddEntry("kf_mks_isr","simulation with ISR","l")
legend.AddEntry("kf_mks_noisr","simulation without ISR","l")
legend.Draw()
cnv.Draw()
cnv.SaveAs("kskpi_isr_noisr.root")
```

```python
cnv = TCanvas()
fl_noisr.kf_data.Draw("kf_chi2>>kf_chi2_noisr(300, 0, 300)", "kf_err==0")
h_noisr = gDirectory.Get("kf_chi2_noisr")
h_noisr.SetLineColor(2)
h_noisr.SetLineStyle(9)
h_noisr.SetLineWidth(3)
h_noisr.SetStats(False)
h_noisr.SetTitle("")
h_noisr.GetXaxis().SetTitle("#chi^2")
h_noisr.GetYaxis().SetTitle("events")
h_noisr.GetXaxis().SetTitleSize(0.06)
h_noisr.GetXaxis().SetLabelSize(0.06)
h_noisr.GetYaxis().SetTitleSize(0.06)
h_noisr.GetYaxis().SetLabelSize(0.06)
fl_isr.kf_data.Draw("kf_chi2>>kf_chi2_isr(300, 0, 300)", "kf_err==0", "same")
h_isr = gDirectory.Get("kf_chi2_isr")
h_isr.SetLineColor(4)
h_isr.SetLineWidth(3)
legend = TLegend(0.1,0.7,0.48,0.9)
legend.AddEntry("kf_chi2_isr","simulation with ISR","l")
legend.AddEntry("kf_chi2_noisr","simulation without ISR","l")
legend.Draw()
cnv.Draw()
cnv.SaveAs("chi2_kskpi_isr_noisr.root")
```

```python
cnv2 = TCanvas()
fl_isr.kf_data.Draw("kf_mks>>kf_mks_isr(512,0.4, 0.6)", "kf_err==0")
h_isr = gDirectory.Get("kf_mks_isr")
h_isr.SetLineColor(4)
h_isr.SetLineWidth(3)
h_isr.SetStats(False)
h_isr.SetTitle("")
h_isr.GetXaxis().SetTitle("m_{#pi^{+}#pi^{#minus}} (GeV)")
h_isr.GetYaxis().SetTitle("events")
h_isr.GetXaxis().SetTitleSize(0.06)
h_isr.GetXaxis().SetLabelSize(0.06)
h_isr.GetYaxis().SetTitleSize(0.06)
h_isr.GetYaxis().SetLabelSize(0.06)
fl_isr_ph.kf_data.Draw("kf_mks>>kf_mks_isr_ph(512,0.4,0.6)", "kf_err==0", "same")
h_isr_ph = gDirectory.Get("kf_mks_isr_ph")
h_isr_ph.SetLineColor(1)
h_isr_ph.SetLineWidth(3)
h_isr_ph.SetLineStyle(7)
legend = TLegend(0.1,0.7,0.48,0.9)
legend.AddEntry("kf_mks_isr","hypothesis without ISR photon","l")
legend.AddEntry("kf_mks_isr_ph","hypothesis with ISR photon","l")
legend.Draw()
cnv2.Draw()
cnv2.SaveAs("kskpi_isr_ph.root")
```

```python
cnv2 = TCanvas()
fl_isr.kf_data.Draw("kf_chi2>>kf_chi2_isr(300,0, 300)", "kf_err==0")
h_isr = gDirectory.Get("kf_chi2_isr")
h_isr.SetLineColor(4)
h_isr.SetLineWidth(3)
h_isr.SetStats(False)
h_isr.SetTitle("")
h_isr.GetXaxis().SetTitle("#chi^{2}")
h_isr.GetYaxis().SetTitle("events")
h_isr.GetXaxis().SetTitleSize(0.06)
h_isr.GetXaxis().SetLabelSize(0.06)
h_isr.GetYaxis().SetTitleSize(0.06)
h_isr.GetYaxis().SetLabelSize(0.06)
fl_isr_ph.kf_data.Draw("kf_chi2>>kf_chi2_isr_ph(300,0,300)", "kf_err==0", "same")
h_isr_ph = gDirectory.Get("kf_chi2_isr_ph")
h_isr_ph.SetLineColor(1)
h_isr_ph.SetLineWidth(3)
h_isr_ph.SetLineStyle(7)
legend = TLegend(0.1,0.7,0.48,0.9)
legend.AddEntry("kf_chi2_isr","hypothesis without ISR photon","l")
legend.AddEntry("kf_chi2_isr_ph","hypothesis with ISR photon","l")
legend.Draw()
cnv2.Draw()
cnv2.SaveAs("chi2_kskpi_isr_ph.root")
```
