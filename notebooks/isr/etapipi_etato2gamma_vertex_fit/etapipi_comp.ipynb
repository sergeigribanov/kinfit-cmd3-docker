```python
from ROOT import TFile, TCanvas, gDirectory, TLegend
```

```python
path_noisr = 'kinfit_etapipi_etato2gamma_run068708.root'
path_isr = 'kinfit_etapipi_etato2gamma_run068707.root'
path_isr_ph = 'isr_photon_run068707.root'
```

```python
fl_noisr = TFile.Open(path_noisr, 'read')
fl_isr = TFile.Open(path_isr, 'read')
fl_isr_ph = TFile.Open(path_isr_ph, 'read')
```

```python
cnv_noisr = TCanvas()
fl_noisr.kf_data.Draw("kf_mgg>>kf_mgg_noisr(256,0.3,0.9)", "kf_err==0")
cnv_noisr.Draw()
```

```python
cnv_isr = TCanvas()
fl_isr.kf_data.Draw("kf_mgg>>kf_mgg_isr(256,0.3,0.9)", "kf_err==0")
cnv_isr.Draw()
```

```python
cnv_isr_ph = TCanvas()
fl_isr_ph.kf_data.Draw("kf_mgg>>kf_mgg_isr(256,0.3,0.9)", "kf_err==0")
cnv_isr_ph.Draw()
```

```python
cnv = TCanvas()
fl_noisr.kf_data.Draw("kf_mgg>>kf_mgg_noisr(256,0.3,0.9)", "kf_err==0")
h_noisr = gDirectory.Get("kf_mgg_noisr")
h_noisr.SetLineColor(2)
h_noisr.SetLineStyle(9)
h_noisr.SetLineWidth(3)
h_noisr.SetStats(False)
h_noisr.SetTitle("")
h_noisr.GetXaxis().SetTitle("m_{#gamma#gamma} (GeV)")
h_noisr.GetYaxis().SetTitle("events")
h_noisr.GetXaxis().SetTitleSize(0.06)
h_noisr.GetXaxis().SetLabelSize(0.06)
h_noisr.GetYaxis().SetTitleSize(0.06)
h_noisr.GetYaxis().SetLabelSize(0.06)
fl_isr.kf_data.Draw("kf_mgg>>kf_mgg_isr(256,0.3, 0.9)", "kf_err==0", "same")
h_isr = gDirectory.Get("kf_mgg_isr")
h_isr.SetLineColor(4)
h_isr.SetLineWidth(3)
legend = TLegend(0.1,0.7,0.48,0.9)
legend.AddEntry("kf_mgg_isr","simulation with ISR","l")
legend.AddEntry("kf_mgg_noisr","simulation without ISR","l")
legend.Draw()
cnv.Draw()
cnv.SaveAs("etapipi_isr_noisr.root")
```

```python
cnv = TCanvas()
fl_noisr.kf_data.Draw("kf_chi2>>kf_chi2_noisr(300, 0, 300)", "kf_err==0")
h_noisr = gDirectory.Get("kf_chi2_noisr")
h_noisr.SetLineColor(2)
h_noisr.SetLineStyle(9)
h_noisr.SetLineWidth(3)
h_noisr.SetStats(False)
h_noisr.SetTitle("")
h_noisr.GetXaxis().SetTitle("#chi^2")
h_noisr.GetYaxis().SetTitle("events")
h_noisr.GetXaxis().SetTitleSize(0.06)
h_noisr.GetXaxis().SetLabelSize(0.06)
h_noisr.GetYaxis().SetTitleSize(0.06)
h_noisr.GetYaxis().SetLabelSize(0.06)
fl_isr.kf_data.Draw("kf_chi2>>kf_chi2_isr(300, 0, 300)", "kf_err==0", "same")
h_isr = gDirectory.Get("kf_chi2_isr")
h_isr.SetLineColor(4)
h_isr.SetLineWidth(3)
legend = TLegend(0.1,0.7,0.48,0.9)
legend.AddEntry("kf_chi2_isr","simulation with ISR","l")
legend.AddEntry("kf_chi2_noisr","simulation without ISR","l")
legend.Draw()
cnv.Draw()
cnv.SaveAs("chi2_etapipi_isr_noisr.root")
```
