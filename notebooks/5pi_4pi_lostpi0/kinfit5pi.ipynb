```python
from run_kinfit import run_kinfit
from draw_funcs import draw_1d_hist, draw_1d_hists
from ROOT import TFile
import matplotlib.pyplot as plt
```

```python
input_path = "/var/kinfit/data/tr_ph_run059700.root"
output_path = "kinfit_run059700.root"
```

```python
run_kinfit(input_path, output_path=output_path)
```

```python
fl = TFile.Open(output_path, 'read')
```

```python
fl.kf_data.Draw("kf_err>>kf_err(3,0,3)", "", "goff")
draw_1d_hist("kf_err", xlabel='error code',
            fraction=0.47,
             info_coords=(0.3, 0.3))
plt.savefig('kf_err_4pi_4pipi0lost.eps')
```

```python
fl.kf_data.Draw("kf_chi2>>kf_chi2(128,0,800)", "kf_err==0", "goff")
draw_1d_hist("kf_chi2", xlabel=r'$\chi^2$',
            fraction=0.47,
             info_coords=(0.3, 0.4))
plt.savefig('kf_chi2_4pi_4pipi0lost.eps')
```

```python
fl.kf_data.Draw("in_m3pi>>in_m3pi(512,0.35,1.75)", "kf_err==0", "goff")
fl.kf_data.Draw("kf_m3pi>>kf_m3pi(512,0.35,1.75)", "kf_err==0", "goff")
draw_1d_hists([('in_m3pi', 'before fit'), ('kf_m3pi', 'after fit')],
              xlabel=r'$M_{\pi^+\pi^-\pi^0}$ (GeV/c$^2$)',
             fraction=0.8,
             info_coords=(0.25, 0.2),
             dh=0.4)
plt.savefig('kf_m3pi_4pipi0lost.eps')
```
