```python
from run_kinfit import run_kinfit
from draw_funcs import draw_1d_hist, draw_1d_hists
from ROOT import TFile
import matplotlib.pyplot as plt
```

```python
input_path = "/var/kinfit/data/tr_ph_run049912.root"
output_path = "kinfit_kskpi_run049912.root"
```

```python
run_kinfit(input_path, output_path=output_path, mfield=1.0)
```

```python
fl = TFile.Open(output_path, 'read')
```

```python
fl.kf_data.Draw("kf_err>>kf_err(3,0,3)", "", "goff")
draw_1d_hist("kf_err", 
             xlabel='error code',
            fraction=0.47,
            info_coords=(0.3, 0.3))
plt.savefig('kf_err_kskpi_4pi.eps')
```

```python
fl.kf_data.Draw("kf_chi2>>kf_chi2(64,0,2500)", "kf_err==0", "goff")
draw_1d_hist("kf_chi2", 
             xlabel=r'$\chi^2$',
             ylabel='events',
            fraction=0.8,
            info_coords=(0.5, 0.5),
            h_y=100, h_x=500)
plt.savefig('kf_chi2_kskpi_4pi.eps')
```

```python
fl.kf_data.Draw("vtx[0]>>kf_vtx_x(64, -1, 1)", "kf_err==0", "goff")
draw_1d_hist("kf_vtx_x", xlabel=r'$x$ (cm)')
plt.savefig('kf_vtx_x_4pi.eps')
```

```python
fl.kf_data.Draw("vtx[1]>>kf_vtx_y(64, -1, 1)", "kf_err==0", "goff")
draw_1d_hist("kf_vtx_y", xlabel=r'$y$ (cm)')
plt.savefig('kf_vtx_y_4pi.eps')
```

```python
fl.kf_data.Draw("vtx[2]>>kf_vtx_z(32, -12, 12)", "kf_err==0", "goff")
draw_1d_hist("kf_vtx_z", xlabel=r'$z$ (cm)')
plt.savefig('kf_vtx_z_4pi.eps')
```
